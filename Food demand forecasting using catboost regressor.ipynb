```python
dd_p = df.groupby(['week','center_id','meal_id']).agg(cum_num_orders = ('num_orders',sum), cum_checkout_price = ('checkout_price',sum), cum_base_price = ('base_price',sum))
dd_p = dd_p.reset_index()
dd_p
```

```python
dd_p['daily_avg_num_orders'] = dd_p['cum_num_orders'] / 7
dd_p['daily_avg_checkout_price'] = dd_p['cum_checkout_price'] / 7
dd_p['daily_avg_base_price'] = dd_p['cum_base_price'] / 7
dd_p
```

```python
df_final = pd.merge(df, dd_p, on=['week','center_id','meal_id'])
df_final
```

```python
x = df_final.drop(['num_orders','id'], axis=1)
y = df_final[['num_orders']]
x
```

```python
x.isnull().sum()
```

```python
y
```

```python
x['center_id'] = x['center_id'].astype(np.object)
x['meal_id'] = x['meal_id'].astype(np.object)
x['city_code'] = x['city_code'].astype(np.object)
x['region_code'] = x['region_code'].astype(np.object)
x['op_area'] = x['op_area'].astype(str)
x['emailer_for_promotion'] = x['emailer_for_promotion'].astype(np.float)
x['homepage_featured'] = x['homepage_featured'].astype(np.float)
x['cum_num_orders'] = x['cum_num_orders'].astype(np.float)
x['week'] = x['week'].astype(np.float)
```

```python
x.dtypes
```

```python
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=0.20,random_state = 1234)
```

```python
categorical_features = np.where(x.dtypes != np.float)[0]
```

```python
categorical_features
```

```python
from catboost import CatBoostRegressor
model=CatBoostRegressor(iterations=50, depth=3, learning_rate=0.1, loss_function='RMSE')
model.fit(train_x, train_y,cat_features=categorical_features,eval_set=(test_x, test_y),plot=True)
```
