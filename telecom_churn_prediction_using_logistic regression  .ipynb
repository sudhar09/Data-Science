```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
```

```python
columns_to_labelencode = ['gender','Partner','Dependents','PhoneService','PaperlessBilling','Churn']
```

```python
df[columns_to_labelencode] = df[columns_to_labelencode].apply(le.fit_transform)
df
```

```python
df.isnull().sum()
```

```python
df = df.fillna(df.mean())
```

```python
df.isnull().sum()
```

```python
df = df.drop(['customerID'],axis=1)
df
```

```python
#x = pd.get_dummies(x,columns=['State'])
df = pd.get_dummies(df)
df
```

```python
x = df.drop(['Churn'],axis=1)
y = df[['Churn']]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.20)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf = LogisticRegression(random_state=0).fit(X_train, y_train)
```

```python
prediction = clf.predict(X_test)
prediction
```

```python
from sklearn import metrics
print(metrics.accuracy_score(y_test, prediction))
```

```python
print('Accuracy of logistic regression classifier on train set: {:.2f}'.format(clf.score(X_train, y_train)))
```

```python
print('Accuracy of logistic regression classifier on test set: {:.2f}'.format(clf.score(X_test, y_test)))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y_test, prediction)
print(confusion_matrix)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, prediction))
```

```python
probs = clf.predict_proba(X_test)
preds = probs[:,1]
fpr, tpr, threshold = metrics.roc_curve(y_test, preds)
roc_auc = metrics.auc(fpr, tpr)
```

```python
rcParams['figure.figsize'] = 10, 7
plt.title('Receiver Operating Characteristic')
plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```
